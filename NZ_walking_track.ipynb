```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster

import geopandas as gpd

```

```python
walking_df = gpd.read_file("/kaggle/input/geojson1/DOC_Walking_Experiences_6475698167360651786.geojson")
walking_df.head()
```

```python
walking_map = folium.Map(location=[-40,174],zoom_start=5)


colour_code= {
    'Easy':'green',
    'Moderate':'orange',
    'Advanced' : 'red'
}

def get_colour_code(level):
    if level in colour_code.keys():
        return colour_code[level]
    else :
        return "blue"

mc = MarkerCluster()

# Your code here
for idx, row in walking_df.iterrows():
        mc.add_child(Marker([row['geometry'].y, row['geometry'].x],
                        icon=folium.Icon(
                            color=get_colour_code(row.difficulty), 
                            icon="person-walking", prefix="fa"),
                               tooltip=row.introduction))

walking_map.add_child(mc)

walking_map
```
